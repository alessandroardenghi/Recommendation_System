```python
import pandas as pd
import numpy as np

from sklearn.metrics.pairwise import cosine_similarity

movie_genres = pd.read_csv('movie_reviews/movie_genres.csv')
user_reviews = pd.read_csv('movie_reviews/user_reviews.csv')
```

```python
user_reviews.drop(['Unnamed: 0','User'], axis=1, inplace=True)
user_reviews.columns = [i for i in range(0,2000)]
user_reviews
```

```python
genre = [g for g in list(movie_genres.columns) if 'genre' in g]
movie_repr = movie_genres[genre]
movie_repr
```

```python
def genre_distance(movie1_repr, movie2_repr):
    return np.sum(np.abs(movie1_repr - movie2_repr))

# Initialize an empty distance matrix
num_movies = movie_repr.shape[0]
distance_matrix = np.zeros((num_movies, num_movies))

# Compute the distance between each pair of movies
for i in range(num_movies):
    for j in range(i, num_movies):
        distance = genre_distance(movie_repr.iloc[i].values, movie_repr.iloc[j].values)
        distance_matrix[i, j] = distance
        distance_matrix[j, i] = distance

distance_matrix
```

```python
distance_matrix = np.abs(movie_repr.values[:, np.newaxis] - movie_repr.values).sum(axis=2)

mean_distance = np.mean(distance_matrix)
max_distance = np.max(distance_matrix)
min_distance = np.min(distance_matrix)
print(mean_distance, max_distance, min_distance)

distance_matrix
```

```python
# Compute Jaccard distance matrix efficiently
A = movie_repr.values[:, np.newaxis]  
B = movie_repr.values

# Compute intersection and union
intersection = np.logical_and(A, B).sum(axis=2)
union = np.logical_or(A, B).sum(axis=2)

# Compute Jaccard similarity)
distance_matrix = (intersection / np.maximum(union, 1))  # Avoid division by zero


mean_distance = np.mean(distance_matrix)
max_distance = np.max(distance_matrix)
min_distance = np.min(distance_matrix)
print(mean_distance, max_distance, min_distance)



distance_matrix
```

```python
neigh_distance = 0.2

def get_nearest_movies(movie_id, neigh_distance):
    mask = distance_matrix[movie_id] >= neigh_distance
    output = np.where(np.array(mask) == 1)[0].tolist()
    return output, len(output)

output, length = get_nearest_movies(18, neigh_distance)
output, length
```

```python
user = 0

candidate_movies = user_reviews.columns[user_reviews.iloc[user] == 0].tolist()
b_u = user_reviews.replace(0, np.nan).mean(axis=0).tolist()
estimated_ratings = {}

for movie_id in candidate_movies:
    output, length = get_nearest_movies(movie_id, neigh_distance)
    N_u = [movie for movie in output if user_reviews.iloc[user, movie] > 0]
    

    if len(N_u) > 0:
        
        estimated_ratings[movie_id] = b_u[movie_id] + np.sum([distance_matrix[movie_id, movie] * (user_reviews.iloc[user, movie] - b_u[movie]) for movie in N_u]) / np.sum([distance_matrix[movie_id, movie] for movie in N_u])
    
    else:

        estimated_ratings[movie_id] = b_u[movie_id]
```

```python


user = 0

# Get candidate movies (movies not rated by the user)
candidate_movies = user_reviews.columns[user_reviews.iloc[user] == 0].to_numpy()

# Compute user-specific bias (replace 0s with NaN before computing mean)
b_u = np.nanmean(user_reviews.replace(0, np.nan), axis=0)

# Dictionary to store estimated ratings
estimated_ratings = {}

# Convert user_reviews to NumPy array for fast indexing
user_ratings = user_reviews.to_numpy()

for movie_id in candidate_movies:
    output, _ = get_nearest_movies(movie_id, neigh_distance)
    
    # Find neighbors that the user has rated
    N_u = np.array([movie for movie in output if user_ratings[user, movie] > 0])

    if N_u.size > 0:
        # Compute weighted sum using NumPy's optimized operations
        numerator = np.dot(distance_matrix[movie_id, N_u], (user_ratings[user, N_u] - b_u[N_u]))
        denominator = np.sum(distance_matrix[movie_id, N_u])

        estimated_ratings[movie_id] = b_u[movie_id] + (numerator / denominator if denominator != 0 else 0)
    else:
        estimated_ratings[movie_id] = b_u[movie_id]  # Default to baseline if no neighbors exist
```
